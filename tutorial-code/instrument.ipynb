```python
import requests
import pandas as pd
import defs
```

```python
session = requests.Session()
```

```python
instrument = "EUR_USD"
count = 10
granularity = "H1"
```

```python
url = f"{defs.OANDA_URL}/instruments/{instrument}/candles"
```

```python
params = dict(
    count = count,
    granularity = granularity,
    prince = "MBA"
)
```

```python
response = session.get(url=url, params=params, headers=defs.SECURE_HEADER)
```

```python
response.status_code
```

```python
data = response.json()
data.keys()
```

```python
instruments = data["instruments"]
```

```python
instrument_data = []
for item in instruments:
    new_ob = dict(
        name = item['name'],
        type = item['type'],
        displayName = item['displayName'],
        pipLocation = item['pipLocation'],
        marginRate = item['marginRate'],
    )
    instrument_data.append(new_ob)
```

```python
for item in instrument_data[:3]:
    print(item)
```

```python
instrument_df = pd.DataFrame.from_dict(instrument_data)
```
